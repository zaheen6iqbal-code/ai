## Loan Eligibility Prediction (Linear Regression):

```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Loan Eligibility Prediction.csv')
df
```

```python
df.info()
```

```python
df.nunique()
```

```python
newdf=df[['Customer_ID','Gender','Married','Dependents','Education','Self_Employed','Applicant_Income','Coapplicant_Income',
          'Loan_Amount','Loan_Amount_Term','Credit_History','Property_Area','Loan_Status']]
newdf
```

```python
newdf.Gender.unique()
```

```python
newdf.Gender=newdf.Gender.replace('Male',0)
```

```python
newdf.Gender=newdf.Gender.replace('Female',1)
newdf
```

```python
newdf.Married=newdf.Married.replace('Yes',1)
```

```python
newdf.Married=newdf.Married.replace('No',2)
newdf
```

```python
newdf.info()
```

```python
Apl_Income=newdf['Applicant_Income']
Apl_Income
```

```python
newdf=newdf.drop(columns=['Applicant_Income'])
```

```python
newdf.info()
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(newdf[['Customer_ID','Gender','Married','Dependents','Coapplicant_Income',
          'Loan_Amount','Loan_Amount_Term','Credit_History']],Apl_Income)
```

```python
reg.coef_
```

```python
reg.fit_intercept
```

```python
reg.intercept_
```

```python
reg.predict([[569,1,2,0,0.0,9,360,1]])
```
