```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('Home Price.csv')
```

```python
df
```

```python
plt.scatter(df.price,df.area,color='blue',marker='+')
```

```python
#y=mx+b    #price = slop*area +Y-intercept

#y=price

#x=area

#m=slope

#Y intercept
df
```

```python
new_df=df.drop('price',axis='columns')
new_df
```

```python
price=df.price
price
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(new_df,price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.fit_intercept
```

```python
reg.predict([[60000]])
```
