```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
pci=pd.read_csv('Canada Per Capita Income.csv')
pci
```

```python
pci.rename(columns={'per capita income (US$)':'per_capita_income_us_dollar'},inplace=True)
pci
```

```python
plt.scatter(pci.per_capita_income_us_dollar,pci.year,color='blue',marker='.')
```

```python
pci2=pci.drop('year',axis='columns')
pci2
```

```python
year=pci.year
year
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(pci2,year)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.fit_intercept
```

```python
reg.predict([[2020]])
```
