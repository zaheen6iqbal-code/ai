```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Car Price.csv')
df
```

```python
df.columns
```

```python
df.rename(columns={'CarName':'Car_Name','carbody':'Car_Body','horsepower':'Horsepower','price':'Price'}, inplace=True)
df
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df.info()
```

```python
newdf=df[['doornumber','Horsepower','compressionratio','curbweight','carheight','carwidth','fueltype','Price']]
```

```python
newdf.doornumber=newdf.doornumber.replace('two',0)
```

```python
newdf.doornumber=newdf.doornumber.replace('four',1)
```

```python
newdf.fueltype.unique()
```

```python
newdf.fueltype=newdf.fueltype.replace('gas',0)
```

```python
newdf.fueltype=newdf.fueltype.replace('diesel',1)
```

```python
newdf
```

```python
newdf.info()
```

```python
Pricedf=newdf['Price']
```

```python
Pricedf
```

```python
newdf=newdf.drop(columns=['Price'])
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(newdf[['doornumber','Horsepower','compressionratio','curbweight','carheight','carwidth','fueltype']],Pricedf)
```

```python
reg.coef_
```

```python
reg.fit_intercept
```

```python
reg.intercept_
```

```python
reg.predict([[1,200,31,5000,70,60,1]])
```

```python
reg.predict([[1,160,8.7,3049,55.5,68.8,0]])
```
