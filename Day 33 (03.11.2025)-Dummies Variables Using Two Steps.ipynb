```python
import pandas as pd
from sklearn import linear_model
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv('Car Price (03.11.2025).csv')
df
```

```python
dumiedf=pd.get_dummies(df['Car Model'])
dumiedf
```

```python
dumiedf=dumiedf.astype(int)
```

```python
merge=pd.concat([df,dumiedf],axis ='columns')
merge
```

```python
df2 = merge.drop(['Car Model','Mercedez Benz C class'],axis='columns')
df2
```

```python
df3 = df2.drop(['Sell Price($)'], axis = 'columns')
df3
```

```python
pri = df2['Sell Price($)']
pri
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df3,pri)
```

```python
df4 = reg.predict(df3)
```

```python
df4
```

```python
reg.score(df3,pri)*100
```
