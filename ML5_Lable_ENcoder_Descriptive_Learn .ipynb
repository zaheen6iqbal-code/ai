```python
import pandas as pd
```

```python
from sklearn import linear_model

```

```python
df = pd.read_csv("XYZ.csv")
```

```python
df
```

```python
dumiesdf=pd.get_dummies(df.town)
dumiesdf = dumiesdf[['monroe township','robinsville','west windsor']].replace({True:1,False:0})
```

```python
merge=pd.concat([df,dumiesdf],axis ='columns')
merge
```

```python
df2 = merge.drop(['town','west windsor'],axis='columns')
```

```python
df3 = df2.drop(['price'], axis = 'columns')
```

```python
pri = df2.price
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df3,pri)
```

```python
df5 = reg.predict(df3)
```

```python
reg.score(df3,pri)*100 #accuracy
```
